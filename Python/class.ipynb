1. Write a `Fraction` class to work with fractions. Let the fraction in our class appear as `numerator/denominator`. A fractional number must be generated by `Fraction(a, b)`, where `a` is the numerator and `b` is the denominator of the fraction.
2. Add the ability to add (add via the addition operator) for the fraction. It is assumed that the operation of addition can be carried out both between fractions and between a fraction and an integer. The result of the operation must be presented as a fraction.
3. Add the ability to take the difference (subtraction through the subtraction operator) for fractions. It is assumed that the subtraction operation can be carried out both for two fractions, and for a fraction and an integer. The result of the operation must be presented as a fraction.
4. Add the ability to multiply (multiply through the multiplication operator) for fractions. It is assumed that the multiplication operation can be carried out both for two fractions, and for a fraction and an integer. The result of the operation must be presented as a fraction.
5. Add the ability to convert a fraction to an integer using the standard `int()` function.
6. Add the ability to convert a fraction to a floating point number using the standard `float()` function.
7. Create a child class `OperationsOnFraction` and add your own methods `getint` and `getfloat` to it, which will return the integer part of the fraction and the floating point representation of the fraction, respectively.
 

```python
def div(a, b):
    while a % b != 0:
        old_num1 = a
        old_num2 = b

        a = old_num2
        b = old_num1 % old_num2
    return b


class Fraction:

    def __init__(self, a, b):
        self.a = a
        self.b = b

    def __str__(self):
        return "{}/{}".format(self.a, self.b)

    def check(self, other):
        if not isinstance(self, Fraction):
            return NotImplemented

        if isinstance(other, Fraction):
            other_a = other.a
            other_b = other.b
        elif isinstance(other, int):
            other_a = other
            other_b = 1
        else:
            return NotImplemented
        return other_a, other_b

    def __add__(self, other):
        other_a, other_b = self.check(other)

        new_a = self.a * other_b + other_a * self.b
        new_b = self.b * other_b
        divider = div(new_a, new_b)

        return Fraction(new_a // divider, new_b // divider)

    def __sub__(self, other):
        other_a, other_b = self.check(other)

        new_a = self.a * other_b - other_a * self.b
        new_b = self.b * other_b
        divider = div(new_a, new_b)

        return Fraction(new_a // divider, new_b // divider)

    def __mul__(self, other):
        other_a, other_b = self.check(other)

        new_a = self.a * other_a
        new_b = self.b * other_b
        divider = div(new_a, new_b)

        return Fraction(new_a // divider, new_b // divider)

    def __int__(self):
        return self.a // self.b

    def __float__(self):
        return self.a / self.b


class OperationsOnFraction(Fraction):

    def getint(self):
        return super().__int__()

    def getfloat(self):
        return super().__float__()


fract = Fraction(2, 6)
fract1 = Fraction(1, 6)

fract2 = OperationsOnFraction(9, 3)
fract3 = OperationsOnFraction(1, 3)

print(fract + fract1)
print(fract - fract1)
print(int(fract * fract1))
print(float(fract * fract1))
print(fract2.getint())
print(fract2.getfloat())
```
