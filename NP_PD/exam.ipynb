## ⭐️⭐️ Question 11

To further analyze the behavior of buyers, we need a data set in which each transaction correctly indicates the price, the number of units of goods (Quantity) and the buyer id (CustomerID). Remove from the table all rows where the price is less than 0 or missing, or the number of units is less than or equal to 0 or missing, or where the customer id is missing.

```python
positive_df = sorted_df_price.loc[(sorted_df_price['UnitPrice'] > 0) & (sorted_df_price['Quantity'] > 0)]
not_null_df = positive_df.dropna(subset=['UnitPrice','Quantity','CustomerID'])
not_null_df
```

## ⭐️⭐️ Question 12

The table for each transaction shows the price per unit of goods (UnitPrice) and the number of units of goods (Quantity). Calculate for each transaction its **total cost** and store it in a new Price column that you add to the table.

What are the minimum and maximum total transaction costs? List separated by commas, rounding to whole numbers.

```python
not_null_df['Price'] = not_null_df['Quantity'] * not_null_df['UnitPrice']
result_df = not_null_df.dropna(how='all')
result_df = result_df.sort_values('InvoiceNo')
result_df.head(10)
```

## ⭐️⭐️ Question 13

In our database, each purchase is represented by one or more transactions. A purchase is uniquely identified by its invoice number (InvoiceNo). Transactions related to the same purchase share the same InvoiceNo.

The cost of the entire purchase is equal to the sum of the total costs of the transactions included in it. Find the cost of the three most expensive purchases. List in descending order, separated by commas, rounded to whole numbers.

```python
grouped_df = result_df.groupby(['InvoiceNo']).sum().sort_values('Price', ascending=False)
grouped_df
```

```python
list_of_price = list(grouped_df.Price.head(3))
for i in list_of_price:
    print(round(i))
```

## ⭐️⭐️ Question 15

Create a new table, purchases, with each row corresponding to a single purchase, with columns
InvoiceNo, InvoiceDate, Price, CustomerID, Country.

Since the InvoiceDate may differ for different transactions within the same purchase, use the earliest of the dates when grouping. Also sort it by date in ascending order.

Please note that InvoiceDate is not a unique field, the data should be additionally sorted by InvoiceNo before taking the first value of the group.


```python
purchases = pd.DataFrame()
purchases = result_df.loc[:, ['InvoiceNo', 'InvoiceDate', 'Price', 'CustomerID', 'Country']]
purchases = purchases.sort_values('InvoiceDate').groupby('InvoiceNo').agg({'InvoiceDate': 'min', 'Price': 'sum', 'CustomerID': 'first', 'Country': 'first'})
purchases
```

## ⭐️⭐️ Question 16

We investigate whether the purchases of our customers are increasing or decreasing after their first purchase on the site. Find the average price of customers' first purchases: for each customer, take their first purchase and average these values. Find the average price of purchases in general. Round these numbers up to integers and list them in that order, separated by commas.

\* If TimeStamps are equal for purchases, the first purchase is considered to be the purchase with the lower InvoiceNo.

```python
mean_first_purchase = purchases.groupby('CustomerID').agg({'Price': 'first'}).mean()
mean_purchase = purchases['Price'].values.mean()
print(mean_first_purchase.values.round(), mean_purchase)
```
