```python
from sympy import *
from sympy.plotting import plot3d

init_printing(use_unicode=False, wrap_line=False, no_global=True)
```

```python
x, y = symbols('x, y')
```

```python
#1
f1 = 2*x + 5*y + 1
f1
```

```python
solve(f1)
```

```python
plot3d(f1)
```

```python
plot(-5*y/2 - 1/2)
```

```python
solve(-5*y/2 - 1/2)
```

```python
#2
f2 = x**2 + y**2 -1
f2
```

```python
solve(f2)
```

```python
plot3d(f2)
```

```python
f_x = sqrt(1-y**2)
f_x
```

```python
plot(f_x, (y, -1, 1))
```
