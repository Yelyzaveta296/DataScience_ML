```python
from sympy import *
from sympy.plotting import plot
from sympy.solvers.inequalities import solve_univariate_inequality

init_printing(use_unicode=False, wrap_line=False, no_global=True)
```

```python
x = Symbol('x')
```

```python
f = x**2 + 3*x - 4
plot(f)
```

```python
f1 = exp(2 - 2*x**2)
f1
```

```python
diff_f1 = (2 - 2*x**2) * exp(2 - 2*x**2)
diff_f1
```

```python
plot(f1, (x, -1, 2.8))
```

```python
solve(diff_f1)
```

```python
solve_univariate_inequality(diff_f1 > 0, x)
```

```python
solve_univariate_inequality(diff_f1 < 0, x)
```

```python
a2, a1, a0 = symbols('a2, a1, a0')
```

```python
MSE = 1/4 * (((a1 * 1 + a0) - 9)**2 + \
             ((a1 * 2 + a0) - 6)**2 + \
             ((a1 * 3 + a0) - 4)**2 + \
             ((a1 * 4 + a0) - 2)**2)
MSE
```

```python
MSEa1 = diff(MSE, a1)
MSEa1
```

```python
MSEa0 = diff(MSE, a0)
MSEa0
```

```python
nonlinsolve([MSEa1, MSEa0], [a1, a0])
```
