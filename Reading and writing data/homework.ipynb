###  10

You may have noticed that the `Temperature` column of all previous dataframes contains a fairly high value. This is because the temperature data is for New York City, and in the US it is customary to measure temperatures in Fahrenheit, not Celsius. The following formula is used to convert temperatures from degrees Fahrenheit to degrees Celsius:

$$
T_C = (T_F-32)\times5/9,
$$

где $T_F$ — temperature in degrees Fahrenheit and $T_C$ is the temperature in degrees Celsius.

Using the above formula, add a new `Temperature_C` column to the merged dataframe from the previous task containing the temperature in degrees Celsius. Temperature data in degrees Fahrenheit is given in the `Temperature` column. Save the resulting dataframe to the `result6.csv` file.


```python
import math 
merged_df['Temperature_C'] = round(((merged_df['Temperature'] - 32) * 5/9), 1)
merged_df.to_csv('Data/result6.csv')
merged_df
```
