For Regression task - to predict phone price, for Classification - to predict os or year, for Clustering - to run clustering algorithm where numbers of clusters and features choose at your discretion.


```python
import numpy as np
import pandas as pd
```

```python
import matplotlib
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D 

matplotlib.rc("font", size=18)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
phones = pd.read_csv('phones.csv')
```

```python
phones.head()
```

```python
fig = plt.figure(figsize=(10, 10))
ax = plt.axes()

ax.scatter(phones['disk'], phones['price'], s = 150)
plt.show()
```

```python
X1 = phones[['disk']]
y1 = phones['price']
```

```python
reg = LinearRegression().fit(X1, y1)
```

```python
reg.predict(X1[0:1])[0]
```

```python
# Classification
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import export_text
```

```python
X2 = phones[['disk', 'price']]
y2 = phones['os']
```

```python
cl = DecisionTreeClassifier().fit(X2, y2)
```

```python
print(export_text(cl))
```

```python
cl.predict(X2[57:58])[0], y2[57]
```

```python
#Clustering
```

```python
fig = plt.figure(figsize=(10, 10))
markers = {"Android" : 'o', 'iOS' : 'v'}

for d, pr, os in zip(phones.disk, phones.price, phones.os):
    plt.scatter(d, pr, s=250, marker = markers[os])
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
X3 = phones[['disk', 'price']]
clust = KMeans(n_clusters=2).fit(X3)
```

```python
[c1, c2] = clust.cluster_centers_
```
