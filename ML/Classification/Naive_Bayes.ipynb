```python
from sklearn.datasets import load_wine
import pandas as pd
```

```python
wine_dataset = load_wine()
wine_dataset
```

```python
wine_df = pd.DataFrame(wine_dataset['data'], columns=wine_dataset.feature_names)
scttr_mtrx = pd.plotting.scatter_matrix(wine_df, c=wine_dataset['target'], figsize=(25, 25))
scttr_mtrx
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(wine_dataset.data[:, 11:13], wine_dataset['target'], random_state=17)
```

```python
nb = GaussianNB()
```

```python
nb_model = nb.fit(x_train, y_train)
```

```python
nb_pred = nb.predict(x_test)
nb_pred
```

```python
accuracy = nb.score(x_test, y_test)
accuracy
```
